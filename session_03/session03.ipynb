User-Defined Functions

```python
#Definig and calling function
def greet():
  print("Hello World!!")
```

```python
greet()
```

Function with Parameters/Arguments

```python
def greet_user(name):
  print(f"Hello, {name}")
```

```python
greet_user("Lalin")
```

```python
def isPalindrome(word):
  return word == word[::-1]

word = input("Enter your word here: ")

if isPalindrome(word):
  print("Yes, It's Palindrome.")
else:
  print("No, It's not palindrome.")
```

Function within a Function / Nested Function

```python
def outer_function():
  print("This is outer function.")
  def inner_function():
    print("This is inner function.")
  inner_function()

outer_function()
```

Iteration: for and while loops

```python
#For loop
fruits = ["apple","banana","orange"]
for f in fruits:
  print(f)
```

```python
#While loop
count = 0
while count < 3:
  print("Count:", count)
  count += 1
```

Iterating Over Dictionaries

```python
person = {"name" : "lalin",
          "age" : 24,
          "is_student" : True}

#Iterate over keys and values
for key, value in person.items():
  print("keys:", key)
  print("values:", value)
```

List comprehensive: A conise way to create lists.
Useful for data processing and transformation.

```python
num = [1,2,3,4,5]
squares = [x ** 2 for x in num]
print(squares)

#Filter even number
even = [n for n in num if n % 2 == 0]
print(even)
```

Error Handling (try, except)

```python
try:
  num = int(input("Enter a number: "))
  print("You entered:", num)
except ValueError:
  print("Invalid ipupt! Please enter a valid number.")
```

Lambda Functions

```python
#Double a number

double = lambda x: x * 2
print("Doubled of 5 is:", double(5))

#Use with map() to apply a functin to a list
num = [1,2,3,4,5]
doubled_num = list(map(lambda x: x * 2, num))
print("Doubled numbers:", doubled_num)
```

Matrix Operation

```python
#Nested Lists (2D)
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
print("Element at row 2, column 3:", matrix[1][2])
```
