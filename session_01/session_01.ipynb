```python
age = 18
if age > 60: #If Evaluated True then the first part will be printed.
    print("You are a senior citizen")
elif age > 18: #If Evaluated True then the second part will be printed.
    print("You are an adult")
else:
    print("You are not an adult")
```

```python
#Dictionary
myDic = {
    "Name" : ["Lalin", "Badhon"],
    "Dept" : ["Cse", "Cse"],
    "Intake" : ["47/2", "47/2"]
}
print(myDic)
```

```python
#Accessing data in dictionary
print(myDic["Name"][0])
print(myDic["Name"][1])
```

```python
#Dictionary update
myDic["Intake"] = "45/2", "46/2"
print(myDic)
```

```python
#Dictionary Update
my_dict = {'brand': 'Ford', 'model': 'Mustang'}
new_data = {'year': 2024, 'color': 'blue'}

my_dict.update(new_data)

print(my_dict)
```

```python
import pandas as pd
pd.DataFrame([my_dict])
```
